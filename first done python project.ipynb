```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\HP\Downloads\Day - 30 - All Practice files\Day - 30 - All Practice files\Day - 30 - Mobile Sales Data.xlsx"

df = pd.read_excel(file_path)

# Clean columns
df.columns = df.columns.str.strip()

print(df.columns)

# Replace with actual column names after checking output
highest = df.loc[df["Units Sold"].idxmax()]

print(highest)

plt.figure(figsize=(8,5))

plt.bar(
str(highest["Date"]),
highest["Units Sold"]
)

plt.title("Highest Units Sold Day")

plt.show()
```

```python
total_units = df["Units Sold"].sum()

print("Total Units Sold Over 30 Days =", total_units)

# Make chart
chart_df = pd.DataFrame({
    "Period": ["30 Days"],
    "Units Sold": [total_units]
})

plt.figure(figsize=(8,5))

sns.boxplot(
    data=chart_df,
    x="Period",
    y="Units Sold"
)

plt.title("Total Units Sold Over 30 Days")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read Excel
file_path = r"C:\Users\HP\Downloads\Day - 30 - All Practice files\Day - 30 - All Practice files\Day - 30 - Mobile Sales Data.xlsx"

df = pd.read_excel(file_path)

# Clean column names
df.columns = df.columns.str.strip()

# Average Units Sold
avg_units = df["Units Sold"].mean()

print("Average Daily Units Sold =", round(avg_units,2))

# Seaborn style
sns.set_style("whitegrid")

# BAR CHART
plt.figure(figsize=(8,5))

sns.barplot(
    x=["Average Daily Sales"],
    y=[avg_units]
)

plt.title("Average Daily Units Sold")
plt.xlabel("Metric")
plt.ylabel("Units Sold")

plt.show()
```

```python
best = (
    df.groupby("Brand")["Units Sold"]
    .sum()
    .sort_values(ascending=False)
)

print(best)

# Seaborn style
sns.set_style("whitegrid")

# Chart
plt.figure(figsize=(10,6))

sns.barplot(
    x=best.values,
    y=best.index
)

plt.title("Best Selling Mobile Brand")
plt.xlabel("Units Sold")
plt.ylabel("Brand")

plt.show()
```
